```python
events = spark.read.csv("/Volumes/workspace/ecommerce/ecommerce_data/2019-Oct.csv", header=True, inferSchema=True)
```

```python
# day-4
delta_path = "dbfs:/Volumes/workspace/ecommerce/ecommerce_data/events"

events.write \
    .format("delta") \
    .mode("overwrite") \
    .save(delta_path)

print("CSV changed to Delta format")
```

```python

# Create managed table
events.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("events_table")

print("managed delta table created ")
```
